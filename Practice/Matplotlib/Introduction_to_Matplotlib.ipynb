## Matplotlib example overflow

```python
#0. import matplotlib.pyplot and get it ready for plotting in Jupyter

import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize =(10,10))  #(Width, Height)

#3. Plot the data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. Save and Show(you save the whole figure)
fig.savefig("sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height= nut_butter_prices.values())
ax.set(title = "Nut Butter Store",
       xlabel = "Butter", 
       ylabel = "Price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Horizontal Bar", 
       xlabel = "Prices", 
       ylabel = "Butter");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist(x);
```

## Two Options for Subplot

```python
# Subplots option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols= 2, figsize = (10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000))
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10,5))

#plot to each different index
ax[0,0].plot(x, x/ 2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### WHich one should you use Pyplot vs matplotlib oo method

* When PLotting something quickly, okay to use pyplot method
* When plotting something more advance, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x = 'age', y = 'chol', c = 'target', figsize=(10,6))
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x = 'age', y = 'chol', c = 'target', ax =ax);

# ax.set_xlim([45, 100])
```
