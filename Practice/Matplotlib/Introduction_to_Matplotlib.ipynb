## Matplotlib example overflow

```python
#0. import matplotlib.pyplot and get it ready for plotting in Jupyter

import matplotlib.pyplot as plt

#1. Prepare data
x = [1,2, 3, 4]
y = [11, 22, 33, 44]

#2. Setup plot
fig, ax = plt.subplots(figsize =(10,10))  #(Width, Height)

#3. Plot the data
ax.plot(x, y)

#4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

#5. Save and Show(you save the whole figure)
fig.savefig("sample-plot.png")
```

## Making figures with Numpy arrays


WE want: 
1) line plot
2) Scatter plot
3) Bar plot
4) Hist plot
5) Subplot


```python
x = np.linspace(0,10,100)
```

```python
#PLot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
#Use the same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
#Another scatter plot
fig, ax =plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height= nut_butter_prices.values())
ax.set(title = "Nut Butter Store",
       xlabel = "Butter", 
       ylabel = "Price");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Horizontal Bar", 
       xlabel = "Prices", 
       ylabel = "Butter");
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist(x);
```

## Two Options for Subplot

```python
# Subplots option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols= 2, figsize = (10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000))
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10,5))

#plot to each different index
ax[0,0].plot(x, x/ 2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
