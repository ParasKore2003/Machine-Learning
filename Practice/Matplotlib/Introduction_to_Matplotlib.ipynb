## Two Options for Subplot

```python
# Subplots option 1

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols= 2, figsize = (10,5))

#plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000))
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2, 
                       figsize=(10,5))

#plot to each different index
ax[0,0].plot(x, x/ 2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### WHich one should you use Pyplot vs matplotlib oo method

* When PLotting something quickly, okay to use pyplot method
* When plotting something more advance, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x = 'age', y = 'chol', c = 'target', figsize=(10,6))
```

```python
# OO Method moxed with Pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x = 'age', y = 'chol', c = 'target', ax =ax);

# ax.set_xlim([45, 100])
```

```python
## OO method from the Scratch
fig, ax = plt.subplots(figsize=(10,6))

#PLot the Data
scatter = ax.scatter(x = over_50['age'], 
                     y=over_50['chol'],
                     c = over_50['target'])
#Customize
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel='Cholestrol')

#Add a legend
ax.legend(*scatter.legend_elements(), title=('Target'));

#Add a horizontal line
ax.axhline(over_50['chol'].mean(), 
           linestyle='--', color = 'magenta')
```

```python
over_50.head()
```

```python
# Subplot of chol, age, thelach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

#Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'])

#Customize ax0
ax0.set(title="Heart Diseases and Cholestrol Level", 
        ylabel= "Cholestrol")

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "target")

#Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle="--")


scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"])

#Customize ax1
ax1.set(title="Heart Disease and Thalach Level",
        xlabel="Age",
        ylabel= "Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title="target")

ax1.axhline(y = over_50["thalach"].mean(), 
            linestyle="--");

#Add a title to fig
fig.suptitle("Heart Diesease Analysis", fontsize = 16, fontweight ='bold')
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

## OO method from the Scratch
fig, ax = plt.subplots(figsize=(10,6))

#PLot the Data
scatter = ax.scatter(x = over_50['age'], 
                     y=over_50['chol'],
                     c = over_50['target'],
                     cmap="winter") #THis changes the color scheme
#Customize
ax.set(title="Heart Disease and Cholestrol Levels", xlabel="Age", ylabel='Cholestrol')

#Add a legend
ax.legend(*scatter.legend_elements(), title=('Target'));

#Add a horizontal line
ax.axhline(over_50['chol'].mean(), 
           linestyle='--', color = 'magenta');
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age, thelach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10),
                               sharex=True)

#Add data to ax0
scatter = ax0.scatter(x = over_50['age'],
                      y = over_50['chol'],
                      c = over_50['target'],
                      cmap = 'winter')

#Customize ax0
ax0.set(title="Heart Diseases and Cholestrol Level", 
        ylabel= "Cholestrol")
ax0.set_xlim([50,80])
ax0.set_ylim([100,600])
#Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = "target")

#Add a mean line
ax0.axhline(y=over_50['chol'].mean(), linestyle="--")


scatter = ax1.scatter(x = over_50["age"],
                      y = over_50["thalach"],
                      c = over_50["target"],
                      cmap = 'winter')

#Customize ax1
ax1.set(title="Heart Disease and Thalach Level",
        xlabel="Age",
        ylabel= "Max Heart Rate")

ax1.set_xlim([50,80])
ax1.set_ylim([60, 200])
ax1.legend(*scatter.legend_elements(), title="target")

ax1.axhline(y = over_50["thalach"].mean(), 
            linestyle="--");

#Add a title to fig
fig.suptitle("Heart Diesease Analysis", fontsize = 16, fontweight ='bold')
```
